```python
#MATPLOTLIB

# Useful for making plots

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10,100) #evenly spaced 100 values between 0-100
y = np.sin(x)
z = np.cos(x)

#print(x)
#print(y)
#print(z)

#Plotting sine wave
plt.plot(x,y)
plt.xlabel('Angle')
plt.ylabel('Sine Values')
plt.title('Sine Wave')
plt.show()

#cosine wave
plt.plot(x,z)
plt.xlabel('Angle')
plt.ylabel('CoSine Values')
plt.title('CoSine Wave')
plt.show()

#print parabola
x = np.linspace(-10,10,20)
y = x**2

plt.plot(x,y,'r.') # Graph will be made with . You can use any any symbol like +
plt.xlabel('values')
plt.ylabel(' Values')
plt.title('Parabola')
plt.show()

x = np.linspace(-5,5,50)
plt.plot(x, np.sin(x), 'g+')
plt.plot(x, np.cos(x), 'r.')
plt.show()


#. BAR PLOT

fig = plt.figure()
ax = fig.add_axes([0,0,1,1]) # 0,0 represents axes and 1,1 represent values
languages = ['English', 'French', 'Spanish' , 'Latin' , 'German']
people = [100,50,150,40,70] # number of people speaking respective above languages
ax.bar(languages,people)
plt.xlabel("Languages")
plt.ylabel("Number of People")
plt.title("Languages Spoken")
plt.show()


# Pie CHART

fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English', 'French', 'Spanish' , 'Latin' , 'German']
people = [100,50,150,40,70] # number of people speaking respective above languages
ax.pie(people, labels=languages, autopct='%1.1f%%') # autopct is required to print % vlaue in piechart
plt.show()

# Scatter Plot

x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)

fig2 =  plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()


#  3D Scatter PLot

#fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()

```
