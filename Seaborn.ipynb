```python
# Seaborn
# Visualization Library

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


#Seaborn has some build in datasets

#total bill vs tip dataset
tips = sns.load_dataset('tips')
print(tips.head())

#Setting a default theme
sns.set_theme()
#Visualize the data
sns.relplot(data=tips, x='total_bill', y='tip', col='time', hue='smoker', style='smoker', size='size')
#hue is the differentiator


```

```python
#Load the iris dataset
iris = sns.load_dataset('iris')
print(iris.head())

#here we will use scatterplot
sns.scatterplot(data=iris,x='sepal_length',y='petal_length',hue='species')
```

```python
#load the titanic dataset

titanic = sns.load_dataset('titanic')
print(titanic.head())

sns.countplot(data=titanic,x='class')
sns.countplot(data=titanic,x='survived')

```

```python
#Bar plot
sns.barplot(x='sex',y='survived', hue='class', data=titanic)
```

```python
#House price prediction from sklearn

#practise distrobution plot
# sns.distplot
# heatmap

boston = pd.read_csv('/content/sample_data/HousingData.csv')
print(boston.head())
sns.distplot(boston['TAX'])
```

```python
#Correlation

correlation=boston.corr()

plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```
